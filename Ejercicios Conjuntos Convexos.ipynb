### <b style = 'color : darkred;'>$\text{Primer Punto:}$</b>

$\textbf{2.1}$ Sea $C\subset\mathbb{R}^{n}$ un conjunto convexo,con $x_{1},\cdots,x_{k} \in C$, y sea $\theta_{1},\cdots,\theta_{k}\in \mathbb{R}$ tal que $\theta_{i}\geq 0$ y $\theta_{1}+\cdots+\theta_{k}=1$. Muestre que $\theta_{1}x_{1}+\cdots+\theta_{k}x_{k}\in C$.$\\\\$

$\textbf{Demostración:}$

Sea $C$ un conjunto convexo, tal que $x_{1},\cdots,x_{k} \in C$ y además $\theta_{1},\cdots,\theta_{k}\in \mathbb{R}$ tales que $\theta_{i}\geq 0$ y $\theta_{1}+\cdots+\theta_{k}=1$, se demostrará que si $z=\theta_{1}x_{1}+\cdots+\theta_{k}x_{k}$ entonces $z\in C$ haciendo inducción matemática sobre $k$:

* $\underline{\text{Caso Base:}}$ Tome $k=2$, entonces $z=\theta_{1}x_{1}+\theta_{2}x_{2}$ con $\theta_1,\theta_2\geq 0$ y además $\theta_1+\theta_2=1$  entonces por ser $C$ convexo se sabe que $z\in C$.$\\\\$

* $\underline{\text{Hipótesis de Inducción:}}$ Supongamos que para $k\in\mathbb{N}$ fijo tal que $k>2$ se cumple que si $z=\theta_{1}x_{1}+\cdots+\theta_{k}x_{k}$ entonces $z\in C$ teniendo $\theta_{i}\geq 0$ y  $\theta_{1}+\cdots+\theta_{k}=1$, entonces se procede a demostrar lo proposición para $k+1$.$\\\\$

* $\underline{\text{Paso Inductivo:}}$ Sean $x_{1},\cdots,x_{k},x_{k+1} \in C$,  y sean $\theta_{1},\cdots,\theta_{k+1}\in \mathbb{R}$ tal que $\theta_{i}\geq 0$ y $\theta_{1}+\cdots+\theta_{k}+\theta_{k+1}=1$, sin perdida de generalidad podemos tomar $\theta_{1}\neq 1$(observe que este procedimiento se puede realizar con cualquier $\theta_{i}\neq 1$), entonces se tiene que:

$$\theta_{1}x_{1}+\cdots+\theta_{k}x_{k}+\theta_{k+1}x_{k+1}=\theta_{1}x_{1}+(1-\theta_{1})(\alpha_{2}x_{2}+\cdots+\alpha_{k}x_{k}+\alpha_{k+1}x_{k+1})$$

$\hspace{7mm}$ En donde $(1-\theta_{1})\alpha_{i}=\theta_{i}$ observe que $\alpha_{i}=\frac{\theta_{i}}{1-\theta_{1}}$, entonces $$\sum_{i=2}^{k+1}\alpha_{i}=\sum_{i=2}^{k+1}\frac{\theta_{i}}{1-\theta_{1}}$$ $\hspace{7mm}$ luego obtenemos lo siguiente:

$$\sum_{i=2}^{k+1}\alpha_{i}=\frac{1}{1-\theta_{1}}\sum_{i=2}^{k+1}\theta_i$$ 

$\hspace{7mm}$ Por hipótesis $\sum_{i=1}^{k+1}\theta_{i}=1$ entonces $\theta_{1}+\sum_{i=2}^{k+1}\theta_{i}=1$, de donde $\sum_{i=2}^{k+1}\theta_i=1-\theta_{1}$, así reemplazando se obtiene finalmente que:$\\\\$ 
$$\sum_{i=2}^{k+1}\alpha_{i}=\frac{1}{1-\theta_{1}}(1-\theta_{1})=1$$

$\hspace{7mm}$ Ahora, se obtiene por hipótesis de inducción que $\alpha_{2}x_{2}+\cdots+\alpha_{k}x_{k}+\alpha_{k+1}x_{k+1}\in C$, de donde, al ser $C$ un conjunto convexo y $\theta_{1}+(1-\theta_{1})=1$ 
$\hspace{7mm}$se tiene que $z=\theta_{1}x_{1}+\cdots+\theta_{k}x_{k}\in C$.

Luego por principio de inducción matemática queda demostrado que $\theta_{1}x_{1}+\cdots+\theta_{k}x_{k}\in C$.$\\\\$

$\textbf{2.3}$ *Convexidad del punto medio*. Un conjunto $C$ es convexo en el punto medio si siempre que dos puntos $a, b$ están en $C$, el punto medio o medio $\frac{(a + b)}{2}$ está en $C$. Obviamente, un conjunto convexo es convexo en el punto medio. Se puede demostrar que, en condiciones suaves, la convexidad del punto medio implica convexidad. (Como caso simple, demuestre que si $C$ es cerrado y el punto medio es convexo, entonces $C$ es convexo.)$\\\\$

$\textbf{Demostración:}$

Sean $a,b\in C$, como $C$ es convexo de punto medio entonces $\frac{a+b}{2}\in C$, de la misma manera tomese $c=\frac{a+\frac{a+b}{2}}{2}=\frac{a}{2}+\frac{a}{4}+\frac{b}{4}=\frac{3a}{4}+\frac{b}{4}$ entonces $c\in C$, ahora tome $c_{1}=\frac{c+\frac{a+b}{2}}{2}=\frac{5a}{8}+\frac{3b}{8}$, es claro que realizando este procedimiento recursivamente un número $n$ de veces entonces se tendrá que $c_{n}=\theta_{n}a+(1-\theta_{n})b$, bastará porbar que $\theta_{n}\to \theta$ cuando $n\to \infty$, con $\theta\in[0,1]$:

Se observa que $0\leq\theta_{n}\leq1$ entonces $\theta_{n}=\alpha_{1}2^{-1}+\alpha_{2}2^{-2}+\cdots+\alpha_{n}2^{-n}$ en donde $\alpha_{i}\in\{1,0\}$, en efecto:

$$\lim_{n\to\infty}(\theta_{n})a+(1-\theta_{n})b=\lim_{n\to\infty} \theta_{n}a+\lim_{n\to\infty} (1-\theta_{n})b=\theta a+(1-\theta)b$$

Como $C$ es cerrado entonces el limite de dicha sucesión debe pertencer al conjunto, por tanto $\theta a+(1-\theta)b \in C$.$\\\\$

$\textbf{2.4}$ Muestre que la envolvente convexa de un conjunto $S$ es la intersección de todos los conjuntos convexos que contienen a $S$. (Se puede usar el mismo método para demostrar que la envolvente cónica, afín o lineal de un conjunto $S$ es la intersección de todos los conjuntos cónicos, o conjuntos afines, o subespacios que contienen $S$.)$\\\\$

$\textbf{Demostración:}$

Sea $\{C_i\}_{i\in I}$ la familia de los conjuntos convexos tales que $S\subset C_i$ para todo $i \in I$, queremos demostrar que $\textbf{conv}(S)=\bigcap_{i\in I}C_i$, para ello usaremos la definición de igualdad entre conjuntos:

* $\subset:$ Observe que si $S\subset C_i$ para todo $i\in I$, entonces $S\subset\bigcap_{i\in I} C_i$, por otro lado, si cada $C_i$ es un conjunto  convexo, entonces $\bigcap_{i\in I} C_i$ es un conjunto convexo que contiene a $S$, por definición $\textbf{conv}(S)$ es el conjunto convexo más pequeño que contiene a $S$, por tanto $\textbf{conv}(S)\subset \bigcap_{i\in I}C_{i}$.

* $\supset:$ Sea $x\in\bigcap_{i\in I}C_i$, por definición de intersección indexada de conjuntos se tiene que $x\in C_i$ para todo $i\in I$, en especial $x\in \textbf{conv}(S)$ pues $\textbf{conv}(S)\in \{C_i\}_{i\in I}$ dado que esta es la familia conformada por todos los conjuntos convexos que contienen a $S$, así $\bigcap_{i\in I}C_i\subset \textbf{conv}(S)$.

Luego por la definición de igualdad entre conjuntos se tiene que $\textbf{conv}(S)=\bigcap_{i\in I}C_i$.$\\\\$

$\textbf{2.5}$ Cual es la distancia entre dos hiperplanos paralelos $\{x \in \mathbb{R}^{n}| a^{T} x = b_{1}\}$ y $\{x \in \mathbb{R}^{n}| a^{T} x = b_{2}\}$?$\\\\$


$\textbf{Solución:}$

Sean $A=\{x \in \mathbb{R}^{n}| a^{T} x = b_{1}\}$ y $B=\{x \in \mathbb{R}^{n}| a^{T} x = b_{2}\}$ dos hiperplanos tales que $A\parallel B$, teniendo en cuenta que $a$ es el vector normal a ambos hiperplanos, entonces se puede medir la distancia entre $A$ y $B$  trazando un hiperplano $C$ que sea paralelo al vector normal $a$, por tanto la distancia entre los puntos de corte entre los hiperplanos $A,C$ y $B,C$ es la medida deseada, supongamos que dicho hiperplano $C$ pasa por el origen, entonces se hallan los puntos de corte de la siguiente manera:

* Puesto que $b_1$ representa el desplazamiento del hiperplano $A$ desde el origen, entonces se obtiene que la distancia del origen a el hiperplano $A$ es $d=\frac{b_1}{\parallel a \parallel_{2}}$, por tanto el punto de corte de $C$ con el hiperpano $A$, se obtiene multiplicando la distancia $d$ con el vector normalizando de $a$, es decir,  $x=\frac{b_{1}}{\parallel a\parallel_{2}}\cdot\frac{a}{\parallel a \parallel_{2}}=\frac{b_{1}}{\parallel a \parallel^{2}_{2}}a$.
* De la misma manera $b_2$ representa el desplazamiento del hiperplano $B$ desde el origen, realizando el mismo procedimiento anterior se tiene que el punto de corte  de $C$ con $B$ es $y=\frac{b_{2}}{\parallel a\parallel_{2}^{2}}a$.

Luego, se halla la distancia del punto $x$ al punto $y$:

$$\begin{align*}
  \parallel x-y\parallel_{2}&=\left|\left| \frac{b_{1}}{\parallel a\parallel_{2}^{2}}a-\frac{b_{2}}{\parallel a\parallel_{2}^{2}}a\right|\right|_{2} &(\text{Definición de $x$ e $y$ })\\
  &=\left|\left| \frac{b_{1}-b_{2}}{\parallel a\parallel_{2}^{2}}a\right|\right|_{2}&\\
  &=|b_1-b_2|\cdot\left|\left|\frac{a}{\parallel a\parallel_{2}^{2}}\right|\right|_{2}&\hspace{20mm}(\text{Propiedades de una norma})\\
  &=\frac{|b_1-b_2|}{\parallel a\parallel_{2}^{2}}\cdot\parallel a\parallel_{2}&\hspace{20mm}(\text{Propiedades de una norma})\\
  &=\frac{|b_1-b_2|}{\parallel a\parallel_{2}}&\\
\end{align*}$$

Así la distancia entre dos hiperplanos paralelos $A$ y $B$ está dada por $\frac{|b_1-b_2|}{\parallel a\parallel_{2}}$.$\\\\$

* $S = \{x \in \mathbb{R}^{n}| x \succeq 0, x^{T}y\leq 1 \text{ para todo } y \text{ con }  ||y||_{2}= 1\}$.


$\textbf{Solución:}$

 El conjunto $S$ se puede definir como la intersección de la bola cerrada en $\mathbb{R}^{n}$ con centro en $0$ y radio $1$, es decir, $B[0,1]$  y el octante positivo de $\mathbb{R}^{n}$ (pues $x\geq 0$), entonces el conjunto $S$ se escribe de la siguiente manera: los puntos $x\in \mathbb{R}^{n}$ tales que $\parallel x\parallel\leq 1$.

  A pesar de que dicho conjunto se escribe como la intersección de dos semiespacios, no es un poliedro pues la bola esta formada por la intersección de infinitos semiespacios y en la definición de poliedro se especifica que debe ser intersección de finitos semiespacios e hiperplanos.$\\\\$

* $S = \{x \in \mathbb{R}^{n}| x \succeq 0, x^{T}y \leq 1\text{ para todo } y \text{ con } \sum_{i=1}^{n} |y_{i}| = 1\}$

$\textbf{Solución:}$

Observe que si $x^{T}y \leq 1$ y además  $\sum_{i=1}^{n} |y_{i}| = 1$ entonces para que se cumpla la desigualdad se tiene que el conjunto $S$ es el mismo conjunto dado por $C=\{x\hspace{1mm}||x_i|\leq 1\}$, entonces en realidad podemos ver este conjunto como la intersección entre el primer octante y $C$, observe que a diferencia de el punto anterior si se toma en $C$ el máximo entonces se va a tener que $\max x_i\leq 1$ entonces se pueden descartar demasiados semiespacios  y se puede formar a $S$ como intersección finita de semiespacios, siendo uno el primer octante y el otro el conjunto $\{x\hspace{1mm}| -1\leq x\leq 1\}$, por tanto $S$ si es un poliedro.$\\\\$

$$ V = \{x \in \mathbb{R}^{n}| \parallel x-x_{0}\parallel_{2}\leq  \parallel x-x_{i}\parallel_{2}, i = 1,\cdots, k\}.$$

$V$ se denomina región de Voronoi alrededor de $x_{0}$ con respecto a $x_{1}, \cdots, x_{k}$.$\\\\$

* Demuestre que $V$ es un poliedro. Exprese V en la forma $V = \{x\hspace{1mm} | Ax\preceq b\}$

  $\textbf{Demostración:}$
  
  En la definición del conjunto $V$, se pide que los $x$ cumplan la siguiente propiedad:
  
  $$\parallel x-x_{0}\parallel_{2}\leq  \parallel x-x_{i}\parallel_{2}$$
  
  sabiendo que la norma de un vector se puede escribir como $\parallel y\parallel=y\cdot y^{T}$ entonces se realiza el siguiente procedimiento:
  
  $$\begin{align*}
  \parallel x-x_{0}\parallel_{2}&\leq  \parallel x-x_{i}\parallel_{2}\\
  (x-x_{0})\cdot(x-x_{0})^{T}&\leq  (x-x_{i})\cdot(x-x_{i})^{T}\\
   (x\cdot x^{T}-x_{0}\cdot x^{T}-x\cdot x_{0}^{T}+x_{0}\cdot x_{0}^{T})&\leq  (x\cdot x^{T}-x_{i}\cdot x^{T}-x\cdot x_{i}^{T}+x_{i}\cdot x_{i}^{T})\\
   (x\cdot x^{T}-2x\cdot x_{0}^{T}+x_{0}\cdot x_{0}^{T})&\leq  (x\cdot x^{T}-2x\cdot x_{i}^{T}+x_{i}\cdot x_{i}^{T})\\
    -2x\cdot x_{0}^{T}+x_{0}\cdot x_{0}^{T}&\leq -2x\cdot x_{i}^{T}+x_{i}\cdot x_{i}^{T}\\
    -2x\cdot x_{0}^{T}+2x\cdot x_{i}^{T}&\leq -x_{0}\cdot x_{0}^{T}+x_{i}\cdot x_{i}^{T}\\
    2x(-x_{0}^{T}+x_{i}^{T})&\leq -x_{0}\cdot x_{0}^{T}+x_{i}\cdot x_{i}^{T}\\
    2x(-x_{0}+x_{i})^{T}&\leq -x_{0}\cdot x_{0}^{T}+x_{i}\cdot x_{i}^{T}\\\\
  \end{align*}$$
  
  Tome $A=(-x_{0}+x_{i})^{T}=\left(\begin{matrix}-x_{0}+x_{1}\\-x_{0}+x_{1}\\ \vdots \\ -x_{0}+x_{k}\end{matrix}\right)$ y $b=-x_{0}\cdot x_{0}^{T}+x_{i}\cdot x_{i}^{T}=\left(\begin{matrix}-x_{0}\cdot x_{0}^{T}+x_{1}\cdot x_{1}^{T}\\-x_{0}\cdot x_{0}^{T}+x_{2}\cdot x_{2}^{T}\\ \vdots \\ -x_{0}\cdot x_{0}^{T}+x_{k}\cdot x_{k}^{T}\end{matrix}\right)$ entonces se puede epresar al conjunto $V$ como $V=\{x\hspace{1mm}|Ax\preceq b\}$, luego $V$ es un poliedro.$\\\\$

$\textbf{2.10}$ Conjunto solución de una desigualdad cuadrática. Sea $C \subset \mathbb{R}^{n}$ el conjunto solución de una desigualdad cuadrática 

$$ C = \{x \in \mathbb{R}^{n}| x^{T} Ax + b^{T}x + c \leq 0\}$$ 

con $A\in S^{n}$, $b\in\mathbb{R}^{n}$ y $c\in\mathbb{R}$.

* Muestre que $C$ es convexo si $A \succeq 0$.

  Se puede probar que $C$ es convexo si la intersección de dicho conjunto con cualquier recta dada por $\{x+tv: t\in\mathbb{R}\} $ da como resultado un conjunto convexo.
  
  Observe que la intersección de $C$ y la recta esta dada por:
  
  $$I=\{x+tv\hspace{1mm}|\hspace{1mm} (x+tv)^{T}A(x+tv)+b^{T}(x+tv)+c\leq 0\text{para }t\in\mathbb{R}\}$$
  
  $$\begin{align*}
     (x+tv)^{T}A(x+tv)+b^{T}(x+tv)+c &=(x^{T}+tv^{T})A(x+tv)+b^{T}x+b^{T}tv+c\\
     &=(x^{T}A+tv^{T}A)(x+tv)+b^{T}x+b^{T}tv+c\\
     &=(x^{T}Ax+x^{T}Atv+tv^{T}Ax+tv^{T}Atv)+b^{T}x+b^{T}tv+c\\
     &=(x^{T}Ax+2x^{T}Atv+tv^{T}Atv)+b^{T}x+b^{T}tv+c\\
     &=(x^{T}Ax+t (2x^{T}Av)+t^{2}(v^{T}Av)+b^{T}x+t(b^{T}v)+c\\
  \end{align*}$$
  
  Si agrupamos los factores de $t^{2}$, $t$ y $t^{0}$ tenemos lo siguiente:
  
   $$ t^{2}(v^{T}Av)+t(2x^{T}Av+b^{T}v)+(c+b^{T}x+x^{T}Ax)$$
   
   Sea $a=v^{T}Av$, $a_1=2x^{T}Av+b^{T}v$ y $a_2=c+b^{T}x+x^{T}Ax$ entonces la expresión anterior es igual a $at^{2}+a_1 t+a_2$, así el conjunto $I$ está dado por:
   
   $$I=\{x+tv\hspace{1mm}|\hspace{1mm} at^{2}+a_1 t+a_2 \leq 0\text{ para }t\in\mathbb{R}\}$$
  
  
  Se observa que este conjunto es convexo siempre que $a>0$ pues si es así la parabola abre hacia arriba y por tanto el interior de dicha parabola tal que sea menor a 0 será en efecto convexa, adicionalmente para que $a>0$ entonces se debe tener que $0\preceq A$.$\\\\$
  
* Demuestre que la intersección de $C$ y el hiperplano definido por $g^{T} x +h = 0$ (donde $g \neq 0$) es convexa si $A + \lambda gg^{T}\succeq 0$ para algún $\lambda \in \mathbb{R}$.

  Sea $P$ el conjunto dado por $P=\{x\hspace{1mm}|\hspace{1mm} g^{T}x+h=0\}$ tomando los $a,a_1$ y $a_2$ que se mencionaron en el anterior ítem, tomemos adicionalmente $a_3=g^{T}v$. $\textbf{2.13}$ *Envolvente cónica de productos exteriores*. Considere el conjunto de productos externos de rango $k$, definido como $C=\{XX^{T} | X \in \mathbb{R}^{n\times k} ,\text{ con } \textbf{rank}(X)= k\}$. Describa su envolvente cónica en términos simples.$\\\\$

$\textbf{Solución:}$ 

Se observa que para un vector $y$ cualquiera $\parallel y\parallel_{2}=yy^{T}$, por tanto al ser una norma se tiene que $yy^{T}\geq 0$, luego por definición del orden ($\preceq$) se tiene que $XX^{T}\succeq 0$, de donde $XX^{T}$ es semidefinida positiva. Adicionalmente, se sabe que $Rank(XX^{T})\leq \min\{rank(X),rank(X^{T})\}$ pero $rank(X)=rank(X^{T})$ por definición de transpuesta de una matriz, luego se tiene que $rank(XX^{T})=rank(X)$ y usando la hipótesis $rank(XX^{T})=k$. Pero, se tiene que $X\in\mathbb{R}^{n\times k}$ de donde $X^{T}\in\mathbb{R}^{k\times n}$ luego el producto de estas matrices pertenece a $\mathbb{R}^{n\times n}$ de donde $k$ al ser el rango de tal multiplicación puede ser a lo sumo $n$.

Sean $A,B\in C$, supongamos que $rank(A+B)=k_1$ tal que $k_1<k$ y sea $V\in \mathbb{R}^{n\times n-k_1}$, supongamos que $V$ es una matriz tal que $rank(V)=N(A+B)$ en donde $N$ es el espacio nulo de la matriz $A+B$, es decir, $V(A+B)=0$.

Como $V(A+B)=0$ entonces $V(A+B)V^{T}=0$ lo que es equivalente a $(VA+VB)V^{T}=0$ y por tanto $VAV^{T}+VBV^{T}=0$, esto implica que $VAV^{T}=0$ y $VBV^{T}=0$ pues $A,B\in C$ y por lo tanto son semidefinidas positivas.

$\bullet$ Para $A$ según el conjunto $C$ existe $X_1\in\mathbb{R}^{n\times k}$ tal que $A=X_1 X_1^{T}$ entonces si $VAV^{T}=0$, se tiene que $V(X_1X_1^{T})V^{T}=0$ de donde $(X_1^{T}V)^{T}X_1^{T}V=0$ esto implica que $X_1^{T}V=0$ y por tanto $X_1X_1^{T}V=0$ de lo que se concluye que $AV=0$, puesto que $B\in C$ se realiza el mismo procedimiento y por tanto se llega a que de igual manera $BV=0$.

De lo anterior, se concluye que si un vector está en el espacio nulo de $A+B$ entonces también está en el espacio nulo de $A$ y de $B$. De donde el espacio nulo tiene en realidad una dimensión de $n-k$ para tanto $A$ como $B$, lo cual implica que el espacio nulo de $A+B$ tiene dimensión $n-k$ y por tanto usando el teorema de rango y nulidad se tiene que en realidad el rango de $A+B$ es $k$.$\\\\$

Se deduce que la envolvente cónica del conjunto de productos externos de rango $k$ es el conjunto de matrices positivas semidefinidas de rango mayor o igual a $k$, junto con la matriz cero.

$\textbf{2.14}$ *Conjuntos ampliados y restringidos*. Sea $S \subset \mathbb{R}^{n}$ , y sea $\parallel · \parallel$ una norma sobre $\mathbb{R}^{n}$.

* Para $a \geq 0$ definimos $S_{a}$ como $\{x\hspace{1mm}|\hspace{1mm}dist(x, S) \leq a\}$, donde $dist(x, S) = inf_{y\in S} \parallel x − y\parallel$. Nos referimos a $S_{a}$ como $S$ expandida o extendida por a. Demuestre que si $S$ es convexa, entonces $S_{a}$ es convexa.$\\\\$

  $\textbf{Solución:}$  

  Sean $x,z\in S_{a}$ y $\theta[0,1]$, además sea $w=\theta x+(1-\theta)z$ entonces se medirá cual es $dist(w,S)$:
  
  $$\begin{align*}
    dist(w,S)&=inf_{y\in S}\parallel w-y\parallel & (\text{Definición de la distancia})\\
    &=inf_{y\in S}\parallel\theta x+(1-\theta)z-y\parallel &( \text{ Definición de $w$})\\
    &=inf_{y\in S}\parallel\theta x+(1-\theta)z-(\theta y_1+(1-\theta)y_2)\parallel &( \text{S es convexo, }y=\theta y_1+(1-\theta)y_2)\\
    &=inf_{y_1,y_2\in S}\parallel\theta x+(1-\theta)z-\theta y_1-(1-\theta)y_2)\parallel &\\
    &=inf_{y_1,y_2\in S}\parallel\theta x-\theta y_1 +(1-\theta)z-(1-\theta)y_2)\parallel &\\
    &=inf_{y_1,y_2\in S}\parallel\theta (x-y_1) +(1-\theta)(z-y_2)\parallel &\\
    &\leq inf_{y_1\in S}\parallel\theta (x-y_1) \parallel +  inf_{y_2\in S}\parallel(1-\theta)(z-y_2)\parallel &(\text{Desigualdad triangular})\\
    &= inf_{y_1\in S}|\theta|\parallel x-y_1 \parallel + |1-\theta| inf_{y_2\in S}\parallel z-y_2\parallel &(\text{Propiedad de la norma})\\
    &= inf_{y_1\in S}\theta\parallel x-y_1 \parallel + inf_{y_2\in S}( 1-\theta )\parallel z-y_2\parallel &(\theta\geq 0)\\
    &= \theta inf_{y_1\in S}\parallel x-y_1 \parallel +( 1-\theta ) inf_{y_2\in S}\parallel z-y_2\parallel &(\theta\geq 0)\\
    &\leq \theta a  +( 1-\theta ) a &(x,z\in S_{a})\\
    &= \theta a  +a-\theta a &\\
    &= a &\\
  \end{align*}$$
  
  Así $dist(w,S)\leq a$ y por tanto se obtiene que $w\in S_{a}$, de donde se concluye que en efecto $S_{a}$ es un conjunto convexo.$\\\\$
  
* Para $a \geq 0$ definimos $S_{−a} = \{x\hspace{1mm}|\hspace{1mm} B(x, a) \subset S\}$, donde $B(x, a)$ es la bola (en la norma $\parallel · \parallel$), centrada en $x$, de radio $a$. Nos referimos a $S_{-a}$ como $S$ contraído o restringido por a, ya que $S_{−a}$ consta de todos los puntos que están al menos a una distancia a de $\mathbb{R}^{n} -S$. Demuestre que si $S$ es convexa, entonces $S_{−a}$ es convexa.$\\\\$

  $\textbf{Solución:}$  
  
  Sean $x,y\in S_{-a}$, entonces por definición de $S_{-a}$ se tiene que  $B(x, a) \subset S$ y $B(y, a) \subset S$, en donde $B(x,a)=\{x_1\hspace{1mm}|\hspace{1mm} \parallel d(x,x_1)\parallel <a \}$, adicionalmente sea $\theta\in[0,1]$ y $w=\theta x+(1-\theta)y$ entonces se tiene que:
  
  $B(w,a)=\{x_2\hspace{1mm}|\hspace{1mm} \parallel w-x_2\parallel <a \}$ esto es los $\parallel (\theta x+(1-\theta)y)-x^{2}\parallel$, veamos si para $x_2\in S$ se cumple que
  
  $\begin{align*}
    \parallel (\theta x+(1-\theta)y)-x_{2}\parallel&=\parallel (\theta x+(1-\theta)y)-(y_1\theta+(1-\theta)y_2)\parallel& (\text{S es convexo})\\
    &=\parallel \theta x+(1-\theta)y-y_1\theta-(1-\theta)y_2)\parallel& (\text{S es convexo})\\
    &=\parallel \theta (x-y_1)+(1-\theta)(y-y_2)\parallel&\\
    &\leq \parallel \theta (x-y_1)\parallel+\parallel(1-\theta)(y-y_2)\parallel&\\
    &=\theta \parallel x-y_1\parallel+(1-\theta)\parallel y-y_2\parallel&\\
    &=\theta a+(1-\theta)a&\\
    &=a&\\
  \end{align*}$
  
  Así $B(w,a)\subset S$ y por definición se concluye que $w\in S_{-a}$, en consecuencia $S_{-a}$ es convexo.$\\\\$

$\textbf{2.15}$ *Algunos conjuntos de distribuciones de probabilidad*. Sea $x$ una variable aleatoria de valor real con $prob(x = a_{i}) = p_{i}$, $i = 1,\cdots, n$, donde $a_{1} < a_{2} < \cdots < a_{n}$. Por supuesto que $p \in\mathbb{R}^{n}$ se encuentra en la probabilidad estándar simplex $P = \{p \hspace{1mm}|\hspace{1mm} 1^{T} p = 1, p\succeq 0\}$. ¿Cuáles de las siguientes condiciones son convexas en $p$? (Es decir, ¿para cuál de las siguientes condiciones es el conjunto de $p \in P$ que satisface la condición  de convexidad?)

*Nota:* Se observa que el conjunto $P$ esta definido por la intersección de un semiespacio y un hiperplano, por tanto $P$ es un poliedro.


* $\alpha \leq E f(x) \leq \beta$, donde $E f(x)$ es el valor esperado de $f(x)$, es decir, $E f(x) =\sum_{i=1}^{n} p_{i}f(a_{i})$. (Se da la función $f :\mathbb{R} \rightarrow\mathbb{R}$.) 
  
  Por definición se tiene que $E x=\sum_{i=0}^{n} a_i \cdot p(X=a_i)$ siendo $X$ la variable aleatoria discreta que tiene como soporte todos los valores $a_i$, adicionalmente $prob(x = a_{i})=p_i$, entonces por la ley del estadístico inconsciente $E f(x)= \sum_{i=0}^{n} f(a_i) \cdot p_i$, por tanto la condición dada anteriormente es equivalente a $\alpha \leq\sum_{i=0}^{n} f(a_i) \cdot p_i \leq \beta$, lo cual es claramente una desigualdad lineal en terminos de las probabilidades de los eventos $a_i$.
  
  

* $prob(x>\alpha)\leq \beta$
  
  Por definición de función de distribución se tiene que $F(a_i)=p(X\leq a_i)$, por tanto por ley del complemento en probabilidad se tiene que $p(x>\alpha)=F(\alpha)-1$ de donde $F(\alpha)-1\leq \beta$, al ser $X$ una variable aleatoria discreta se tiene que $F(\alpha)=\sum_{\alpha_{i}\leq \alpha} p(X=\alpha_{i})$ entonces reemplazando tenemos que $\sum_{\alpha_{i}\leq \alpha} p(X=\alpha_{i})-1\leq \beta$, de donde $\sum_{\alpha_{i}\leq \alpha} p(X=\alpha_{i})\leq \beta+1$, lo cual es de igual manera una desigualdad lineal en terminos de las probabilidades $p_i$.
  
  

* $E|x^{3}|\leq\alpha E|x|$
  Observe que usando la ley del estadístico inconsiente se tiene que $E|x|=\sum_{i=0}^{n} |a_i|p(X=a_i)=\sum_{i=0}^{n} |a_i|p_i$ y $E|x^{3}|=\sum_{i=0}^{n} |a_i^{3}|p(X=a_i)=\sum_{i=0}^{n} |a_i^{3}|p_i$, entonces tenemos que 
  
  $$E|x^{3}|\leq\alpha E|x| \text{ es equivalente } E|x^{3}|-\alpha E|x|\leq0$$
  
  Reemplazando los valores obtenidos para las esperanzas halladas, se tiene lo siguiente:
  
  $$\begin{align}\sum_{i=0}^{n} |a_i|p_i -\alpha\sum_{i=0}^{n} |a_i^{3}|p_i&\leq 0\\
  \sum_{i=0}^{n} |a_i|p_i -\sum_{i=0}^{n}\alpha|a_i^{3}|p_i&\leq 0\\
  \sum_{i=0}^{n} |a_i|p_i -\alpha|a_i^{3}|p_i&\leq 0\\
  \sum_{i=0}^{n} (|a_i|-\alpha|a_i^{3}|)p_i&\leq 0\\
  \end{align}$$
   
  Si se mira esta desigualdad en términos de los $p_i$ entonces en efecto se tiene una desigualdad lineal, dando como resultado un poliedro.



* $Ex^{2}\leq\alpha$
  
   Usando la ley del estadístico inconsiente se tiene que $Ex^{2}=\sum_{i=0}^{n} (a_i^{2})p(X=a_i)=\sum_{i=0}^{n} (a_i^{2})p_i$ entonces tenemos que 
  
  $$Ex^{2}\leq\alpha \text{ es equivalente } \sum_{i=0}^{n} (a_i^{2})p_i\leq\alpha$$
  
  Si se mira esta desigualdad en términos de los $p_i$ entonces en efecto se tiene una desigualdad lineal, dando como resultado un poliedro.
  


* $Ex^{2}\geq\alpha$
  
   Usando la información suministrada en el anterior ítem se tiene que 
  
  $$Ex^{2}\geq\alpha \text{ es equivalente } \sum_{i=0}^{n} (a_i^{2})p_i\geq\alpha$$
  
  Por tanto, esta desigualdad en términos de los $p_i$ entonces en efecto se tiene una desigualdad lineal, dando como resultado un poliedro.



* $var(x)\leq \alpha$ donde $Var(x)=E(x-Ex)^{2}$ es la varianza de $x$.

  Observe que $Var(x)=E(x-Ex)^{2}$, pero la varianza de una variable aleatoria también se puede escribir en terminos de su esperanza de la siguiente manera:
  
  $$Var(x)=Ex^{2}-(Ex)^{2}$$ 
  
  Sabiendo que $Ex^{2}=\sum_{i=0}^{n} (a_i^{2})p_i$ y que $Ex=\sum_{i=0}^{n} (a_i)p_i$, entonces se tiene que
  
  $$Var(x)=\sum_{i=0}^{n} (a_i^{2})p_i-(\sum_{i=0}^{n} (a_i)p_i)^{2}$$ 
  
  Como la condición pide que $var(x)\leq \alpha$ entonces $\sum_{i=0}^{n} (a_i^{2})p_i-(\sum_{i=0}^{n} (a_i)p_i)^{2}\leq \alpha$, esto no es convexo pues tomando los valores $n=3$, $a_1=0$, $a_2=1$ y $a_3=1$ se tiene que escogiendo a $\alpha \frac{1}{5}$ para las siguientes probabilidades $p^{(1)}=(1,0,0)$ y $p^{(2)}=(0,1,0)$ cumplen la condición dada pero $p(\frac{1}{2},\frac{1}{2},0)$ no lo hace pues
  
  $$var(x)=0\cdot\frac{1}{2}+ 1\cdot\frac{1}{2}-(0\cdot\frac{1}{2}+ 1\cdot\frac{1}{2})^{2}=\frac{1}{2}-\frac{1}{4}=\frac{1}{4}$$
  
  Este resultado es mayor que el alfa escogido.
  
  

* $var(x)\geq \alpha$

  Usando lo dicho anteriormente se tiene que $$Var(x)=\sum_{i=0}^{n} (a_i^{2})p_i-(\sum_{i=0}^{n} (a_i)p_i)^{2}\geq \alpha$$, esto se puede reescribir de una manera mas conveniente:
  
  Sea $b=(a_1^{2},a_2^{2},\cdots,a_n^{2})$ y tome a $A=aa^{t}$ entonces $\sum_{i=0}^{n} (a_i^{2})p_i=b^{T} p$, adicionalmente $(\sum_{i=0}^{n} (a_i)p_i)^{2}=p^{T}Ap$ , por lo tanto $b^{t}p-p^{T}Ap\geq \alpha$ es en efecto un conjunto convexo pues $A$ es semidenfinida positiva.
  
  
 
* $cuantil(x)\geq \alpha$ donde $cuantil(x)= inf\{\beta | prob(x \leq \beta) \geq0.25\}$
  
  Por definición de cuantil $cuantil(x)\geq \alpha$ es lo mismo que escribir $prob(x \leq \beta) <0.25$ para todo $\beta<\alpha$.Como los valores soporte de la variable aleatoria $X$ son los $a_i$ se debe buscar el mayor $a_i$ menor a $\alpha$ para el cual $prob(x \leq a_i)< 0.25$ de aqui $\sum_{k=0}^{i} p_k<0.25$, como es un menor estricto entonces esto define un semiespacio abierto, pues no contiene a su frontera. 


* $cuantil(x)\leq \alpha$

  Por definición de cuantil $cuantil(x)\leq\alpha$ es lo mismo que escribir $prob(x \leq \beta) \geq0.25$ para todo $\beta\geq\alpha$, ahora si $prob(x \leq \beta) \geq0.25$ entonces se tiene que si $i$ es el minimo tal que $F(x\leq a)\leq 0.25$ entonces se tiene que por propiedades de la probabilidad $\sum_{k=i+1}^{n}p_{k}\geq 0.25$ lo cual es una desigualdad lineal en terminos de $p_i$.

$\textbf{2.16}$ Demuestre que si $S_{1}$ y $S_{2}$ son conjuntos convexos en $\mathbb{R}^{m+n}$, entonces también lo es su suma parcial

$$\begin{align*}
\theta(x,y_{1}+y_{2})+(1-\theta)(x_1,y_{3}+y_{4})&=(\theta x,\theta(y_{1}+y_{2}))+((1-\theta)x_1,(1-\theta)(y_{3}+y_{4}))\\

&=(\theta x+(1-\theta)x_1,\theta y_{1}+\theta y_{2}+(1-\theta)y_{3}+(1-\theta)y_{4})\\

\end{align*}$$

Como $S_1$ y $S_2$ son convexos entonces se tiene que $ (\theta x+(1-\theta)x_1,\theta y_{1}+(1-\theta)y_{3})\in S_1$ y además $(\theta x+(1-\theta)x_1,(\theta y_{2}+(1-\theta)y_{4}))\in S_2$. $\textbf{2.18}$ Funciones lineales-fraccionarias y conjuntos convexos. Sea $f : \mathbb{R}^{m} \rightarrow \mathbb{R}^{n}$ la función lineal-fraccional

$$f(x) = \frac{(Ax + b)}{(c^{T}x + d)} \hspace{5mm} dom f = \{x | c^{T}x + d > 0\}.$$

Suponga la matriz: 

$$Q=\left(\begin{matrix}A&b\\c^{T}&d\end{matrix}\right)$$

es no singular. Demuestre que f es invertible y que $f^{-1}$ es una aplicación fraccionaria lineal. Dé una expresión explícita para $f^{-1}$ y su dominio en términos de $A, b, c$ y $d$. Insinuación. Puede ser más fácil expresar $f^{-1}$en términos de $Q$.$\\\\$

$\textbf{Solución:}$ 

Observe que si  a cada punto $x$ de $\mathbb{R}^{n}$ se le asocia un rayo dado por $P(x)=\{t(x,1)\hspace{1mm}|\hspace{1mm} t>0\}$ el cual se encuentra en $\mathbb{R}^{n+1}$, ahora, como $t>0$ la ultima componente del rayo es positiva, luego para tomar el rayo desde el origen se agrega la condición $t\geq 0$, como esta correspondencia entre los elementos de $\mathbb{R}$ y los rayos con ultima componente postiva es una biyección entonces podemos escribir:

$$f=P^{-1}(Q(P(x))$$

Esto se puede realizar pues si tomamos un $x\in Dom f$ se tiene que $c^{T}x + d >0$ entonces tomando el rayo asociado a $x$ por medio de $P$, la matriz $Q$ o mejor llamada transformación lineal actual sobre el rayo para producir otro rayo que seria $QP$, entonces para recuperar de nuevo a la función F es necesario componer por $P^{-1}$.

Luego la inversa de la función $f$ se puede escribir como:

$$f^{-1}=P^{-1}(Q^{-1}(P(x))$$

En donde $Q^{-1}$ es una transformación lineal proyectiva que se encuentra asociada a $f^{-1}$.$\\\\$

$\textbf{Algoritmo con semiespacios:}$ Sea $C$ un conjunto finito de puntos en $\mathbb{R}^{n}$.

1. Sean $a,b\in C$, se traza el segmento de recta de $x$ a $y$ dado por $\theta x+(1-\theta)y$, siendo $\theta\in[0,1]$.

2. Se realiza el anterior paso con todas las posibles combinaciones de elementos en $C$.
3. Se obtiene una figura $D$ en la cual todos los puntos de $C$ estan conectados por una recta (se podria considerar esta como un grafo completo de los puntos de $C$).



4. Se halla el vector normal $c$ a la recta en la cual esta contenida el segmento de recta, para así formar el semiespacio de cada uno de los segmentos $c^{T}x\leq y$ o $c^{T}x\geq y$.



5. Se mira cuales puntos de  $C$ estan en los semiespacios formados para cada segmento.





6. Si algún segmento tiene un semiespacio en el cual no se encuentran puntos de $C$ entonces este segmento hace parte de la frontera de $D$.
7. Se encuentran todos los segmentos frontera de $D$.
8. La unión de todos los segmentos frontera de $D$ la cual llamaremos $F$, será la envolvente convexa de $C$.




9. Teniendo la envolvente convexa, se tranzan los rayos de los puntos $z\in C$ tales que $z\in F$ hacia el origen.
10. La figura formada será en efecto el cono convexo de los puntos de $C$.$\\\\$




$\textbf{Algoritmo Alterno:}$ Para $C\subset \mathbb{R}^{2}$ tales que los puntos de $C$ esten ubicados en el primer cuadrante podemos calcular el cono convexo de la siguiente forma:

* Se toma un punto $x\in C$ y se forma el rayo $\theta x$ en donde $\theta>0$.
* Se realiza el procedimiento anterior para cualquier punto $y\in C$.
* Cuando se tenga la totalidad de los rayos, se toma el rayo con mayor pendiente y se traza el semiespacio de puntos menores a el.
* Se toma el rayo con menor pendiente y se traza el semiespacio de puntos mayores a el.
* Se toma la intersección de los dos semiespacios anteriores.
* La intersección dará como resultado el cono convexo de $C$. $\\\\$

### <b style = 'color : darkred;'>$\text{Tercer Punto:}$</b>

Verificar que el conjunto de soluciones de un conjunto de desigualdades de matrices lineales es convexa.

$\textbf{Demostración:}$

Se debe mostrar que el conjunto $C=\{x\hspace{1mm}| f(x)\preceq B\}$ es un conjunto convexo, en donde $f(x)=x_{1}A_{1}+x_{2}A_{2}+\cdots+x_{n}A_{n}$, para $A_{i},B \in S^{m}$ donde $S^{m}$ es el conjunto de matrices de tamaño $m\times m$ que son simetricas.

Sean $x,y\in C$ y sea $\theta\in[0,1]$, tomese $z=\theta x+(1-\theta)y$ entonces:

$$f(z)=z_{1}A_{1}+z_{2}A_{2}+\cdots+z_{n}A_{n}\text{ en donde  cada } z_{i}\text{ corresponde  a } \theta x_{i}+(1-\theta)y_{i} $$  

Esto es, aprovechando las propiedades de acosiatividad en la multiplicación de matrices y agrupando:

$\begin{align*}
  f(z)&=(\theta x_{1}+(1-\theta)y_{1})A_{1}+(\theta x_{2}+(1-\theta)y_{2})A_{2}+\cdots+(\theta x_{n}+(1-\theta)y_{n})\\
  &=(\theta x_{1}A_{1}+(1-\theta)y_{1}A_{1})+(\theta x_{2}A_{2}+(1-\theta)y_{2}A_{2})+\cdots+(\theta x_{n}A_{n}+(1-\theta)y_{n}A_{n})\\
   &=(\theta x_{1}A_{1}+\theta x_{2}A_{2}+\cdots+\theta x_{n}A_{n})+((1-\theta)y_{1}A_{1}+(1-\theta)y_{2}A_{2}+\cdots+(1-\theta)y_{n}A_{n})\\
   &=\theta(x_{1}A_{1}+x_{2}A_{2}+\cdots+x_{n}A_{n})+(1-\theta)(y_{1}A_{1}+y_{2}A_{2}+\cdots+y_{n}A_{n})\\
\end{align*}$

Pero $x,y\in C$ entonces $f(x)=x_{1}A_{1}+x_{2}A_{2}+\cdots+x_{n}A_{n}\preceq B$ y $g(y)=y_{1}A_{1}+y_{2}A_{2}+\cdots+y_{n}A_{n}\preceq B$ entonces:

$$\begin{align*}
  f(z)&=\theta f(x)+(1-\theta)f(y)\\
  &\preceq \theta B+(1-\theta) B\\
  &=\theta B+B-\theta B\\
  &=B\\
\end{align*}$$

Por tanto $f(z)\preceq B$ y por tanto queda demostrado que $z\in C$, luego $C$ es un conjunto convexo.
